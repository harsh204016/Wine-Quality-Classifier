```python
from sklearn.datasets import load_wine
import pandas as pd
from pycaret.classification import *
```

```python
 data = load_wine(as_frame=True)
```

```python
df = data.data
```

```python
df.head()
```

```python
label = data.target
print(label.value_counts())

d = {'quality':label}
label = pd.DataFrame(d)
label.head()
```

```python
dataset = pd.concat([df,label],axis=1)
dataset.head()
```

```python
dataset.drop(['proline','od280/od315_of_diluted_wines'],inplace=True,axis=1)
dataset.shape
```

```python
model_classify = setup(dataset,'quality',session_id=45)
```

```python
%%time
best = compare_models()
```

```python
et_model = create_model('et')
```

```python
evaluate_model(et_model)
```

```python
predict_model(et_model)
```

```python
save_model(et_model, model_name = 'extra_tree_model')
```
